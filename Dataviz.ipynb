```python
iris = load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
iris_df['target'] = iris_df['target'].astype(int) # Convert target column to integer
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5)

pd.plotting.parallel_coordinates(iris_df, 'target')
```

##pandas : scatter กระจัดกระจาย matrix

##color

###color map

```python
matplotlib.rcParams['figure.figsize'] = (5, 5)
pd.plotting.parallel_coordinates(iris_df, 'target', color=('green','red','blue'))
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = (5, 5)
pd.plotting.parallel_coordinates(iris_df, 'target', color=('c', 'm', 'y', 'k'))
plt.show()
```
