### เชื่อมใบขับขี่กับคนว่าเป็นใคร

```python
# Merge person_df and license_df on license_id and id
person_license_df = pd.merge(person_df, license_df, left_on='license_id', right_on='id', how='inner')

# Display the first few rows of the merged dataframe
display(person_license_df.head())
```

##หาผู้ที่ขับรถ Tesla ผมสีแดง

```python
# Filter person_license_df for individuals who drive a Tesla and have red hair
tesla_red_hair_individuals = person_license_df[
    (person_license_df['car_make'] == 'Tesla') &
    (person_license_df['hair_color'] == 'red')
]

# Display the filtered data
display(tesla_red_hair_individuals)
```

##บงการเป็นผู้หญิง สูงประมาณ 5'5" ถึง 5'7"

```python
# Filter the tesla_red_hair_individuals by gender (female) and height (between 65 and 67 inches)
potential_masterminds = tesla_red_hair_individuals[
    (tesla_red_hair_individuals['gender'] == 'female') &
    (tesla_red_hair_individuals['height'].between(65, 67))
]
# Display the filtered individuals with selected columns
display(potential_masterminds[['name', 'age', 'height', 'hair_color', 'gender', 'car_make', 'car_model']])
```
