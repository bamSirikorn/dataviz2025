# Dataset ใหม่ ที่มี ตัวเลข

````
movie_df = pd.read_csv('/content/drive/MyDrive/data_viz_2024_DATA/IMDb Movies India.csv', encoding='latin-1')
print(movie_df.shape)
movie_df.head()
````

````
# prompt: ถ้า dropna ข้อมูลเหลือกี่ %

remaining_percentage = (movie_df.dropna().shape[0] / movie_df.shape[0]) * 100
print(f"Remaining data after dropna: {remaining_percentage:.2f}%")
````

````
movie_df = movie_df.dropna()
````

````
# prompt: ตัด ( ) ออกจากค่าใน column Year และแปลงค่าเป็นตัวเลข

# ตัด ( ) ออกจากค่าใน column Year และแปลงค่าเป็นตัวเลข
movie_df['Year'] = movie_df['Year'].str.replace('(', '').str.replace(')', '').astype(int)
````

````
movie_df.head()
````

````
movie_df['Duration'] = movie_df['Duration'].str.replace(' min', '').astype(int)
````

````
movie_df
````

### ```.describe()``` คำนวณค่าทางสถิติของข้อมูลที่เป็นตัวเลข

````
movie_df.describe()
````

````
# prompt: สร้างตารางของ​ Director ที่มี Name เยอะที่สุด เอาแค่ข้อมูลของคนที่เยอะที่สุด

# นับจำนวนหนังที่แต่ละ Director กำกับ
director_counts = movie_df['Director'].value_counts()

# หา Director ที่กำกับหนังเยอะที่สุด
most_prolific_director = director_counts.index[0]
num_movies = director_counts[0]

# สร้างตารางของ Director ที่กำกับหนังเยอะที่สุด
most_prolific_director_df = movie_df[movie_df['Director'] == most_prolific_director]

# แสดงตาราง
print(f"Director with most movies: {most_prolific_director} ({num_movies} movies)")
most_prolific_director_df
````
